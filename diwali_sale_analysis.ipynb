Diwali Sale Analysis

Data cleaning

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Diwali Sales Data.csv",encoding='unicode-escape') #used to avoid encoding error
```

```python
df.shape # to display rows & columns
```

```python
df.head(10) # to display top 10 rows
```

```python
df.info() # display info for basic analysis here we know that column status and unnamed1 are empty
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True) # to drop the empty columns
# axis=1 means complete row
```

```python
pd.isnull(df) # shows which cell is null
```

```python
pd.isnull(df).sum() # shows how many values are null in each column
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int') # float to int
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Married'}) # to rename and not saved
```

```python
df.describe()
```

```python
df.columns
```

```python
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gender)
```

```python
age = sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in age.containers:
    age.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).tail(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

```python
mar = sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(5,7)})
for bars in mar.containers:
    mar.bar_label(bars)
```

```python
sales_mar = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_mar,hue='Gender')
```

```python
occ = sns.countplot(data=df,x='Occupation')
sns.set(rc={'figure.figsize':(25,10)})
for bars in occ.containers:
    occ.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=sales_occ)
```

```python
cat = sns.countplot(data=df,x='Product_Category')
sns.set(rc={'figure.figsize':(30,10)})
for bars in cat.containers:
    cat.bar_label(bars)
```

```python
sales_cat = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_cat)
```

```python
sales_cat = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).tail(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_cat)
```

Married females spend the most.
People of age group 26-35 spend the most.
Uttar Pradesh, Maharashtra and Karnataka have highest orders and money spent while Rajsthan, Punjab and Telangana have least.
Highest money was spent by people working in IT, Healthcare and Aviation.
Highest money was spent on Food, Clothing & Apparel and Electronic Gadgets and least on Hand & Power Tools, Veterinary and Office.
