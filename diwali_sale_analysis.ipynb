Data cleaning

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Diwali Sales Data.csv",encoding='unicode-escape') #used to avoid encoding error
```

```python
df.shape # to display rows & columns
```

```python
df.head(10) # to display top 10 rows
```

```python
df.info() # display info for basic analysis here we know that column status and unnamed1 are empty
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True) # to drop the empty columns
# axis=1 means complete row
```

```python
pd.isnull(df) # shows which cell is null
```

```python
pd.isnull(df).sum() # shows how many values are null in each column
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int') # float to int
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Married'}) # to rename and not saved
```

```python
df.describe()
```

```python
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```
